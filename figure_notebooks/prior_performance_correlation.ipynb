```python
import pandas as pd
import seaborn as sns
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
```

```python
gsd_evaluation_file = '/Users/julian/temp/BAYESIAN_SKIP_MODEL_EVALUATION/geneva_stroke_dataset/convolutional_bayesian_skip/test_evaluation/test_stats.csv'
gsd_eval_df = pd.read_csv(gsd_evaluation_file)
```

```python
r, p = pearsonr(gsd_eval_df["dice_prior"], gsd_eval_df["dice_les"])

sns.set_style("whitegrid")
gsd_fig = plt.figure(figsize=(5, 5))
ax = gsd_fig.add_subplot(1,1,1)
sns.regplot(x="dice_prior", y="dice_les", data=gsd_eval_df, marker='+')
ax.set_xlabel('Dice Score of Prior')
ax.set_ylabel('Dice Score of Model')
ax.set_title('Geneva Stroke Dataset')
ax.legend(title='Pearson’s correlation', loc='upper left', labels=[f'r={r:.2f}, p={p:.1e}'])
```

```python
gsd_fig.savefig('../static/figures/prior_model_perf_corr/gsd_prior_model_performance_corr.svg', format='svg')
```
